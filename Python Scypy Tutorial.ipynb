### **Matplotlib**

````python
%matplotlib inline
import matplotlib.pyplot as plt
````

````python
x = [0, 1, 2]
y = [0, 1, 4]
y
````

````python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(x, y)
plt.show()
````

````python
x_highres = np.linspace(0, 2, 20)
y_highres = x_highres ** 2
````

````python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x, y)
axes.plot(x_highres, y_highres)

plt.show()
````

````python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres, color='red', linestyle='dashed', linewidth=3, marker='o', 
          markerfacecolor='blue', markersize=5)
````

````python
fig = plt.figure(figsize=(12,8))
axes = fig.add_subplot(111)
axes.plot(x_highres, y_highres, 'r--')
axes.set_title('$y=x^2$')
axes.grid()
axes.set_xlabel('$\theta$')
axes.set_ylabel('y')
````
